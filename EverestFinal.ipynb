```
import matplotlib.pyplot as plt
from skimage.exposure import match_histograms
from skimage import io
```

```
src = io.imread("/content/2001.jpg")
src2 = io.imread("/content/2020.jpg")
ref = io.imread("/content/2007.jpg")

matched = match_histograms(src, ref, multichannel=True)
matched2 = match_histograms(src2, ref, multichannel=True)
```

```
io.imsave('/content/2001match.jpg', matched)
```

```
io.imsave('/content/2020matched.jpg', matched2)
```

```
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows=1, ncols=5, figsize=(8, 3),sharex=True, sharey=True)
for aa in (ax1, ax2, ax3, ax4, ax5): 
    aa.set_axis_off()

ax1.imshow(src)
ax1.set_title('Src, 2001')
ax2.imshow(ref)
ax2.set_title('Ref, 2007')
ax3.imshow(matched)
ax3.set_title('2001/2007 Match')
ax4.imshow(src2)
ax4.set_title('Src2, 2020')
ax5.imshow(matched2)
ax5.set_title('2007/2020 Match')

plt.tight_layout()
plt.show()
```
